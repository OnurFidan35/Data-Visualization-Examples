**Veri Seti Hikayesi**

```python
import seaborn as sns
fmri=sns.load_dataset("fmri") #Beyne yerleştirilen bir cihaz aracılığıyla toplanan sinyallerin oluşturduğu veri seti
df=fmri.copy()
df.head()
```

```python
df.shape
```

```python
df["timepoint"].describe()
```

```python
df["signal"].describe()
```

```python
df.groupby("timepoint")["signal"].count()
```

```python
df.groupby("signal").count() #Her bir timepointe denk gelen sinyallerin farklı olduğunu 1064 satır olduğundan ve her
# sinyala 1 timepoint denk geldiğinden doğrulamış olduk.
```

```python
df.groupby("timepoint")["signal"].describe()
```

**Çizgi Grafiği**

```python
sns.lineplot(x="timepoint",y="signal",data=df);
```

```python
sns.lineplot(x="timepoint",y="signal",hue="event",data=df); #Sıçramanın sebebinin stim eventi olduğunu görebiliyoruz.
```

```python
sns.lineplot(x="timepoint",y="signal",hue="event",style="region",data=df); #Kırılımı arttırınca aslında pik yapma sebebinde region değişkeninin de 
#etkisi olduğunu görüyoruz.
```
