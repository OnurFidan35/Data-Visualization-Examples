# Histogram ve Yoğunluk Grafikleri

```python
import seaborn as sns
diamonds=sns.load_dataset("diamonds")
df=diamonds.copy()
df.head()
```

```python
sns.distplot(df.price, kde=False) #kde parametresi yoğunluğu da çizme
```

```python
#Histogram grafiği bize sayısal verilerin dağılımı hakkında bilgi verir
df.price.describe()
```

```python
sns.distplot(df.price,bins=10,kde=False)
```

```python
#bins parametresi ile kaç adet sütun olacağını belirtiriz 10 verdiğimizde mesela; 0 ile 2500 arasında 25000 adet fiyat olduğunu görüyoruz
```

```python
sns.distplot(df.price)
```

```python
#Bu şekilde yoğunluk grafiğini de çizdirmiş olduk. Grafiği incelediğimizde standart sapmanın çok düşük olmadığını ve 
#Çarpıklık söz konusu olduğunu görüyoruz.
#Histogram ve yoğunluk aynı anda gösterilmek istendiğinde y ekseninde ölçekler birbirinden farklı olacağından dolayı
#seaborn arka tarafta ölçeği olasılık yoğunluk fonksiyonun ölçeğine göre gerçekleştirip bu ölçeğe göre indirgiyip sunuyor.
```

```python
sns.distplot(df.price,hist=False,) #Sadece yoğunluk grafiği istersek.
```

```python
sns.kdeplot(df.price, shade=True)
```

```python
(    sns.FacetGrid(df,
               hue="cut",
               height=5,
               xlim=(0,10000) #Belirgin farklar ve değişimler 0-10000 aralığında olduğu için x eksenini bu aralıkta aldık.
              
            ).map(sns.kdeplot,"price",shade=True).add_legend()    )   #FacetGrid grafik üstüne eklenen boyutları bölerek göstermek için kullanılır.
                                                                #Yukarıdaki gibi bir yoğunluk grafiği oluşturup bunun üzerine boyut eklemek istiyorsak
                                                                #Bu durumda 2 grafiği mapliyoruz yani eşleştiriyoruz. 
```

Ideal sınıfını en kaliteli olarak kabul etmemize rağmen 
burada çok ilginç bir şekilde Ideal sınıfının 0 ile 2000 aralığında yoğunlaştığını görüyoruz.

```python
sns.catplot(x="cut",y="price",hue="color",kind="point",data=df);
```

Ideal cut sınıfının altında color kategorik değişkeninin farklı sınıflarının farklı değerler oluşturmuş olması colorun cut ile kesiştirildiğinde price değişkenine göre ayırt edici bir bilgi taşıdığını gösteriyor.

Y değişkenine bakarsak aggragate edildiği ve fiyatların ortalama değerleri ve bunların etrafındaki aşağı-yukarı yönlü sapmalar ifade edilmiş.
