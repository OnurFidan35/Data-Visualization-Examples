# Korelasyon Grafikleri

```python
import seaborn as sns
tips=sns.load_dataset("tips")
df=tips.copy()
df.head()
```

**Scatterplot**

```python
sns.scatterplot(x="total_bill",y="tip",data=df); #İki değişken arasındaki ilişkiyi ifade eder.
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="time",data=df);
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="day",data=df);
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="day",style="time",data=df); #Kırılımları arttırarak daha fazla bilgi elde ediyoruz.
```

```python
sns.scatterplot(x="total_bill",y="tip",size="size",data=df); #Sürekli değişkenler için size kullanıyoruz
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="size",size="size",data=df); #Daha da belirginleşti.
```

**Doğrusal İlişikinin Gösterilmesi**

```python
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x="total_bill",y="tip",data=df);
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",data=df);
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",col="time",data=df);
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",col="time",row="sex",data=df);
```
